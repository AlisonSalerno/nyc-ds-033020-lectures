```python
rolling_stones = [] #list should be under function
import csv
with open("data.csv") as f:
        reader = csv.DictReader(f)
        for row in reader:
            rolling_stones.append(row)
```

```python
def findAlbumInfo(name): #find by name function #(our_data,name)?
    for item in rolling_stones:
        if item['album'] == name:
            return item
    else:
        return None
```

```python
def findByRank(num): #find by rank
    for item in rolling_stones:
        if item['number'] == num:
            return item
    else:
        return None
```
